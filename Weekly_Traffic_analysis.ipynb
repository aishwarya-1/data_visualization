```python
import string
for index,row in df['Date'].iteritems():
    df['Date'].loc[index] = df['Date'].loc[index].replace('-', '/')
```

```python
li = {}
for index,row in df['Date'].iteritems():
    if(df['Date'].loc[index][1]=='1'):
        li[(int(df['Date'].loc[index][3])*10+int(df['Date'].loc[index][4]))] = []
    elif(df['Date'].loc[index][1]=='2'):
        li[(int(df['Date'].loc[index][3])*10+int(df['Date'].loc[index][4]))+31] = []
```

```python
for index,row in df.iterrows():
    sum = 0
    #print(index, row[7])
    for i in range(8, 32):
        sum = sum + row[i]
    avg = math.ceil(sum/24)
    if(row[7][1]=='1' and (int(row[7][3])*10+int(row[7][4]))>=20):
        li[(int(row[7][3])*10+int(row[7][4]))].append(avg)
        #print(sum)
    elif((int(row[7][3])*10+int(row[7][4]))+31 <50):
        li[(int(row[7][3])*10+int(row[7][4]))+31].append(avg)
        #print(sum)
```
