#### Finging the average traffic for each street.

```python
li = ['12:00-1:00 AM','1:00-2:00AM', '2:00-3:00AM','3:00-4:00AM', '4:00-5:00AM','5:00-6:00AM', '6:00-7:00AM','7:00-8:00AM',
      '8:00-9:00AM','9:00-10:00AM','10:00-11:00AM','11:00-12:00PM','12:00-1:00PM', '1:00-2:00PM', '2:00-3:00PM', '3:00-4:00PM',
      '4:00-5:00PM', '5:00-6:00PM','6:00-7:00PM','7:00-8:00PM','8:00-9:00PM', '9:00-10:00PM', '10:00-11:00PM', '11:00-12:00AM']
y = {}
```

```python
def street_average(st_id):
    x = []
    for i in li:
        sum = 0
        ctr = 0
        avg = 0
        for index, rows in df[i].iteritems():
            if(df['GIS ID'].loc[index]==st_id):
                sum = sum + df[i].loc[index]
                ctr = ctr+1;
        avg = math.ceil(sum/ctr)     
        x.append(avg)
    y[st_id] = x
    #print(y[st_id])
```

```python
streets = set(df['GIS ID'])
streets
for i in streets:
    print(i)
    street_average(i)
```

```python
len(list(y.values()))
```

```python
traffic_11 = np.asarray(list(y.values()))
traffic_11 = traffic_11.sum(axis=0)
```

```python
traffic_11 = traffic_11/378
```

```python
x = li
#y1 = list(y1.values())
#y2 = list(y2.values())
#year = ['2011-2012', '2012-2013']

di = pd.DataFrame(dict(Hour=x, Traffic_Volume=traffic_11))

sns.set_context('poster')
sns.factorplot(x = "Hour", y="Traffic_Volume", data=di, kind="bar",palette="Blues_d",size=8,aspect=2,legend_out=False)
plt.title("Traffic_Volume 2011")
plt.xticks(rotation=90)
```
