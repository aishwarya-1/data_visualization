```python
df1 = df1.dropna()
df1['GIS ID'] = df1['GIS ID'].astype(int)
df1[li] = df1[li].astype(int)
```

### WORKING PART

```python
li = ['12:00-1:00 AM','1:00-2:00AM', '2:00-3:00AM','3:00-4:00AM', '4:00-5:00AM','5:00-6:00AM', '6:00-7:00AM','7:00-8:00AM',
      '8:00-9:00AM','9:00-10:00AM','10:00-11:00AM','11:00-12:00PM','12:00-1:00PM', '1:00-2:00PM', '2:00-3:00PM', '3:00-4:00PM',
      '4:00-5:00PM', '5:00-6:00PM','6:00-7:00PM','7:00-8:00PM','8:00-9:00PM', '9:00-10:00PM', '10:00-11:00PM', '11:00-12:00AM']
y1 = {}
y2 = {}
```

```python
def street_average(st_id, col, y, df):
    m = []
    x = df.loc[df[col] == st_id]
    x = pd.DataFrame(x)
    #print(x['Segment ID'])
    for i in li:
        sum = 0
        ctr = 0
        avg = 0
        for index, rows in x[i].iteritems():
            sum = sum + x[i].loc[index]
            ctr = ctr+1;
        avg = math.ceil(sum/ctr)    
        m.append(avg)
    y[st_id] = m
```

```python
streets1 = list(set(df1['GIS ID']))
streets2 = list(set(df2['Segment ID']))
for i in streets1:
    street_average(i, 'GIS ID', y1, df1)
for i in streets2:
    street_average(i,  'Segment ID', y2, df2)    
```
